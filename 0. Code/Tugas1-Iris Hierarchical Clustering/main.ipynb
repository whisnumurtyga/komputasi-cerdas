```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Muat dataset bunga iris
iris = pd.read_csv('Iris.csv')
iris = iris.iloc[:, 1:5]
print(iris)
```

```python
# Hitung matriks jarak
from scipy.spatial import distance_matrix
dist_matrix = distance_matrix(iris, iris)
dist_matrix
```

```python
# Lakukan hierarchical clustering
linkage_matrix = linkage(dist_matrix, method='ward')  # 'ward' linkage umumnya digunakan untuk data numerik
dendrogram(linkage_matrix, labels=np.arange(1, 151), truncate_mode='level', p=3)  # Label dari 1 hingga 150
plt.title("Dendrogram Hierarchical Clustering - Iris Dataset")
plt.xlabel("Jumlah Data")
plt.ylabel("Jarak")
plt.show()
```
